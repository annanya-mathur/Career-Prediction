```python
import pandas as pd
```

```python
df=pd.read_csv('/content/exams (2).csv')
df.head(5)
```

# **CART(Classification and Regression Tree) ALGORITHM IS USED HERE**
> **For the tree all the dataset will go to root node**

```python
from sklearn.model_selection import train_test_split
```

```python
Independent=df[['Total']]           # X
Dependent=df[['race/ethnicity']]    # y 
x_train,x_test,y_train,y_test=train_test_split(Independent,Dependent,test_size=0.3)              # 70 % of the records will go to trained data and other to test data
```

```python
x_train.info()
```

```python
y_train.info()
```

```python
x_test.info()
```

```python
y_test.info()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

# **Creating Instance From Decision Tree Classifier**




```python
dct=DecisionTreeClassifier()
dct
```

# **Fitting Our Model**

```python
dct.fit(x_train,y_train)
dct
```

```python
y_pred=dct.predict(x_test)
y_pred
```

# **Confusion Matrix**

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```
